```python
import pandas as pd
import time
import re
import numpy as np
import nltk
import requests
import random
from collections import defaultdict
from bs4 import BeautifulSoup
from nltk.sentiment.vader import SentimentIntensityAnalyzer
```

# Name the stocks and the dates

```python
days = np.linspace(1,31,31)
dates = []
for d in days:
    dates.append("2023-10-" + str(round(d)) if d//10 else "2023-10-0" + str(round(d)))
dates.append('2023-11-01')
companies = ["DIS", "XOM"] # list of companies
headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/119.0.0.0 Safari/537.36 Edg/119.0.0.0', 'Connection':'close'}
```

```python
# Define the URL of the article on Yahoo.com
# MSFT site:yahoo.com daterange:2460219-2460249 -AVI -FLV -WMV -MOV -MP4
_dict = defaultdict(list)
for company in companies:
    for j in range(len(dates) - 1):
        url = f"https://www.google.com/search?q={company} site:yahoo.com after:{dates[j]} before:{dates[j+1]} -AVI -FLV -WMV -MOV -MP4 -jpg -jpeg -png"
        response = requests.get(url, headers=headers)
        soup = BeautifulSoup(response.content, "html.parser")
        links = soup.find_all("a")
        count = 0
        for link in links:
            try:
                if link['href'][:30] == 'https://finance.yahoo.com/news' or link['href'][:26] == 'https://www.yahoo.com/news':
                    _dict[(company, dates[j])].append(link["href"])
                    count += 1
                if count == 2: # number of articles
                    break
            except:
                pass
        t = random.uniform(10,15)
        time.sleep(t)
```

```python
results = defaultdict(set)
for k, v in _dict.items():
    print(k)
    for url in v:
        print(url)
        response = requests.get(url, headers=headers)
        if response.status_code == 200:
            print("yes")
            try:
                # Parse the page content using BeautifulSoup
                soup = BeautifulSoup(response.content, "html.parser")

                # Find the article content based on the specific HTML tags
                article_content = soup.find('div', class_='caas-body').find_all('p')

                # Extract the text from the article content
                extracted_texts = []
                for paragraph in article_content:
                    extracted_texts.append(paragraph.get_text())
                extracted_text = " ".join(extracted_texts)

                # Initialize the VADER sentiment analyzer
                sia = SentimentIntensityAnalyzer()

                # Analyze sentiment of the extracted text
                sentiment_scores = sia.polarity_scores(extracted_text)
                if results[k]:
                    print(k)
                    for name, score in sentiment_scores.items():
                        results[k][name] = results[k][name]+score
                        results[k][name] = round(results[k][name], 4)
                        print(name,score)
                   # results[k] = results[k]/len(v)
                else:
                    results[k] = sentiment_scores
            except:
                pass
        print(results)
        t = random.uniform(1,3)
        time.sleep(t)
    results[k]['neg']= round(results[k]['neg']/ len(v),4)
    results[k]['compound']= round(results[k]['compound']/ len(v),4)
    results[k]['neu']= round(results[k]['neu']/ len(v),4)
    results[k]['pos']= round(results[k]['pos']/ len(v),4)
```

```python
# Convert the dictionary_items to a dictionary with nested dictionaries
data_dict = {key: value for key, value in results.items()}
#del data_dict[0]
#del data_dict[1]
#del data_dict[2]
```

```python
# Create a DataFrame from the dictionary
df = pd.DataFrame.from_dict(data_dict, orient='index')
```

```python
# Reset index to have a more standard DataFrame structure
df.reset_index(inplace=True)
df.rename(columns={'level_0': 'Stock', 'level_1': 'Date'}, inplace=True)

# Display the DataFrame
print(df)
df.to_csv('sentiment_scores_DisXom.csv')
```
